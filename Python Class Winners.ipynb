```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_doc, 'html.parser')

print(soup.prettify())
```

```python
# Initialize a set of strings
ENTRANTS_SET = {'zero2sudo'}

# For every a tag, add a unique username into the entrants SET
for link in soup.find_all('a'):
    # ex. user_link = /@zero2sudo
    user_link = link.get('href')
    
    # Get the username by stripping `/@` 
    # from all links before we add it to the set
    # user_link[2:] means "discard the first 2 elements of the string",
    # the first two elements being '/' and '@'
    username = user_link[2:]
    ENTRANTS_SET.add(username)
    
ENTRANTS_SET

# Remove me because I don't need to be in the drawing
ENTRANTS_SET.remove('zero2sudo')
ENTRANTS_SET
```

```python
# Convert the set to a list, so we can shuffle the list
ENTRANTS_LIST = list(ENTRANTS_SET)
# List of entrants, before we shuffled the deck
ENTRANTS_LIST
```

```python
# Import the random library
import random

# Shuffle the deck of entrants
random.shuffle(ENTRANTS_LIST)
# Display the shuffled deck
ENTRANTS_LIST
```
